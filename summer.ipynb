## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df
```

```python
df_gold = df[df['Medal'] == 'Gold']
```

```python
data = []

for sport in df_gold['Sport'].unique():
    data.append([sport, len(df_gold[df_gold['Sport'] == sport])])
pd.DataFrame(data,columns = ['Sport','freq']).sort_values(by='Sport',ascending=False).plot(x='Sport',y='freq',kind='bar',figsize = (15,20))
```

## 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for sport in df['Sport'].unique():
    data.append([sport, len(df[df['Sport'] == sport])])
pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Sport',ascending=False).plot(x='Sport',y='Medal',kind='bar',figsize = (15,20))
```

## 4. Which player has won most number of medals? (Top 5)

```python
data = []

for player in df['Athlete'].unique():
    data.append([player, len(df[df['Athlete'] == player])])
pd.DataFrame(data,columns = ['Athlete','freq']).sort_values(by='Athlete',ascending=False).plot(x='Athlete',y='freq',kind='bar',figsize = (50,80))
```
